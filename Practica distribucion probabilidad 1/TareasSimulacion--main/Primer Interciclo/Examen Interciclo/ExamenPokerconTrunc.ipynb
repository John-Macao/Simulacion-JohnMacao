```python
import random
import numpy as np
import matplotlib.pyplot as plt
import collections
import math

diferentesEi=0.3024*100
parEi=0.5040*100
parparEi =0.1080*100
terciaEi=0.0720*100
terciaparEi=0.0090*100
pokerEi=0.0045*100
quintillaEi=0.0001*100


lista = []
semilla =  1032022  #Se borro el cero inicial ya que no permite su ejecucion 
random.seed(semilla)  
for _ in range(100):
    lista.append(random.random())


#print(lista)
#Resultado

mult_list = [i * 100000 for i in lista]
#print(mult_list)
round_list = [ (math.trunc(num)) for num in mult_list]
print(round_list)





diferentes=0
par=0
parpar =0
tercia=0
terciapar=0
poker=0
quintilla=0

#print(list)


for xn in range(len(round_list)):
    #num = list[xn]
    num=round_list[xn]
    #print(num)
    x = [str(a) for a in str(num)]
    #print(len(x))
    while len(x) < 5:
        x.insert(0, '0')
        #print(x)


    counter=collections.Counter(x)
    # counter.pop('.')

    # if counter['0'] == '1':
    #     counter.pop('0')
    # else:
    #     counter['0'] = counter['0']-1

    #print(counter)


    if len(counter) ==5:
        #Diferentes
        diferentes=diferentes+1
    elif len(counter) ==4:
        #Par
        par=par+1
    elif len(counter) ==3:
        #Tercia, par par 
        if 3 in counter.values():
            tercia = tercia +1
        else:
            parpar=parpar+1
    elif len(counter) ==2:
        #poker, tercia y par 
        if 4 in counter.values():
            poker=poker+1
        else:
            terciapar=terciapar+1
    elif len(counter) ==1:
        #quintilla
        quintilla=quintilla+1

print("======")
print("Diferentes: " + str(diferentes))
print("Par: " + str(par))
print("Par Par: " +str(parpar))
print("Tercia: " + str(tercia))
print("TerciaPar: " +str(terciapar))
print("Poker: "+ str(poker))
print("Quintilla: " + str(quintilla))


#Aplicacion de formula

fdiferentes=((diferentesEi - diferentes)**2)/diferentesEi
fpar=((parEi - par)**2)/parEi
fparpar =((parparEi - parpar)**2)/parparEi
ftercia=((terciaEi- tercia)**2)/terciaEi
fterciapar= ((terciaparEi-terciapar)**2)/terciaparEi
fpoker=((pokerEi-poker)**2)/pokerEi
fquintilla=((quintillaEi-quintilla)**2)/quintillaEi



print("======")
print("Diferentes: " + str(fdiferentes)) #33
print("Par: " + str(fpar))  #52
print("Par Par: " +str(fparpar)) #7
print("Tercia: " + str(ftercia)) #
print("TerciaPar: " +str(fterciapar))
print("Poker: "+ str(fpoker))
print("Quintilla: " + str(fquintilla))

print("======")
print("Sumatoria:" + str(fdiferentes+fpar+fparpar+ftercia+fterciapar+fpoker+fquintilla)) #1.72

estado=""

if (fdiferentes+fpar+fparpar+ftercia+fterciapar+fpoker+fquintilla) > 12.59:
    estado="Rechazado"
else:
    estado="Aceptado"

print("Comparacion Chi-Cuadrada para ver si cumple: "+ estado)

```
