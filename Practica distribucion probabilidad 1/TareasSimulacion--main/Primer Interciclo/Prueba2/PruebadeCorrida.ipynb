#### 1.- Random de Python (01032022)

```python
import random
import numpy as np
import matplotlib.pyplot as plt

lista = []
semilla =  1032022  #Se borro el cero inicial ya que no permite su ejecucion 
random.seed(semilla)  
for _ in range(100):
    lista.append(random.random())
#print(lista)

npa = np.asarray(lista, dtype=np.float32)

B = np.where(npa >= 0.5, 1, 0).transpose()
    
#print(B)
C0=0     
for x in range(len(B)-1):
    #print ("Actual:" + str(x))
    if B[x] == 1:
        #print ("Comparado:" + str(B[x+1]))
        if B[x+1] == 0:
            C0=C0+1
    elif B[x]== 0:
         #print ("Comparado:" + str(B[x+1]))
        if B[x+1] == 1:
            C0=C0+1

if (B[len(B)-1]) != (B[len(B)-2]):
    C0=C0+1

print ("C0 = " + str(C0))

#Contamos los Ceros y unos
unique, counts = np.unique(B, return_counts=True)

n0=np.asarray((counts[0] )).T
n1=np.asarray((counts[1] )).T

print ("n0= "+str(n0))
print ("n1= "+str(n1))
#Sabemos que el total de datos es 100

n=100
print("Total de numeros= " + str(n))

#Ahora utilizamos las formulas y remplazamos

UC0 = ((2*(n0)*(n1))/n) + (1/2)
print("UC0= " + str(UC0))

o2C0 = (2*(n0)*(n1))*((2*(n0)*(n1))-n)/((n**2)*(n-1))
print("o2C0= " + str(o2C0))

Z0 = (C0-UC0)/np.sqrt(o2C0)
print("Z0= " + str(Z0))


numeros = [n0,n1]
nombres = ["0","1"]
plt.pie(numeros, labels=nombres)
plt.show()
```
