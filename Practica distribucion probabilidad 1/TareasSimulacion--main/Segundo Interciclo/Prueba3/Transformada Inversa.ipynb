```python
import random
import math
import numpy as np
import matplotlib.pyplot as plt
```

**Problema:**

Se ha descubierto una sustancia extraña que ayuda a la regeneracion de globulos rojos, pero, extrañamente a veces causa Hemólisis y para poder medir su comportamiento ha medido el numero de globulos rojos que nacen o mueren segun la cantidad de la sustancia (Ri) insetada en cada una de las 100 muestras.

#### Simulacion:

```python
# Crear lista de numeros random y sus resultados
lista = []
#listaRes = []
semilla =  653324
random.seed(semilla)  
print('Muestra', '\t ri', '\t Celulas')

for i in range(100):
    u = random.random()
    lista.append(u)
    x = math.tan( math.pi * ( u - ( 1 / 2)) )
    #listaRes.append(x)
    print(str(i +1) , '\t' +str(round(u,7)), '\t' +str(round(x,7)) )
```

#### Gráficas y Resultados:
**Gráfica de Función de densidad**

```python
listaOrd = sorted(lista)
listaResOrd=[]

for i in listaOrd:
    x = 1 / (math.pi * (1 + i**2))
    listaResOrd.append(x)

plt.plot(listaOrd, listaResOrd)
plt.title('Distribución Cauchy: Función de densidad (Lado derecho de la campana)') 
# El lado izquierdo necesita valores negativos y Ri no posee ninguno
plt.ylabel('Globulos R.')
plt.xlabel('Cantidad de sustancia')
plt.show()
```

**Gráfica de Función de probabilidad acumulada**

```python
listaResOrd=[]

for i in listaOrd:
    x = math.tan( math.pi * ( i - ( 1 / 2)) )
    listaResOrd.append(x)

plt.plot(listaOrd, listaResOrd)
plt.title('Distribución Cauchy: Función acumulativa')
plt.ylabel('Globulos R.')
plt.xlabel('Cantidad de sustancia')
plt.show()
```
