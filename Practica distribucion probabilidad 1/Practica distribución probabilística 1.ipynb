```python
import numpy as np
import pandas as pd 
from scipy import stats
import math
import matplotlib.pyplot as plt
import seaborn as sns
```

## Analisis descriptivo 1

```python
root = './fifa_datos.csv'
data = pd.read_csv(root, header=0, delimiter=',')
data = data.drop(['Unnamed: 0'], axis=1)
data.head()
```

## Descripción

```python
data.describe()
```

## Correlación

```python
data.corr()
```

## Mapa de calor

```python
sns.heatmap(data.corr())
```
