# Ejercicio en clase

1 Con los datos de Fifa, organizar a los jugadores de acuerdo al peso en las siguientes escalas y generar un cuadro tipo PIE

- Debajo 125 Lbs.
- 125-150
- 150-175
- 175 o superior

```python
fifa = "fifa_datos.csv"
data = pd.read_csv(fifa, delimiter=',')
data
```

```python
#Delimitando datos

data["Weight"]= data["Weight"].replace({"lbs":""}, regex=True) 
data = data.dropna(subset=["Weight"])
data["Weight"] = data["Weight"].astype(int)


debajo = data.loc[data["Weight"] < 125].count()[0]

medio = data.loc[(data["Weight"] >= 125) & (data["Weight"] < 150)].count()[0]

medio2 = data.loc[(data["Weight"] >= 150) & (data["Weight"] < 175)].count()[0]

superior = data.loc[data["Weight"] >= 175].count()[0]
```

```python
#Figura

plt.figure()

ets = ["Debajo 125 Lbs", "125-150", "150-175", "175 o superior"]
plt.title("Comparacion del pesos de los Jugadores")
plt.pie([debajo, medio, medio2, superior], labels = ets, autopct='%.2f %%')

plt.show()

```
