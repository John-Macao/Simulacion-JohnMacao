```python
#Librerias
#Librerias
import plotly.express as px
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel('Conformación de los Grupos Económicos y su Comportamiento Tributario (4).xlsx', sheet_name='RESUMEN DE GRUPOS ECONÓMICO', header=4)
df.head(5)
```

## 1. Generar gráfico de barras de solo las empresas que han mejorado y/o empeorado su ranking con respecto al año anterior.

```python
data1 = pd.DataFrame(df, columns=['Variación Ranking Tamaño 2020 - 2021'])

data1.head(10)
```

```python

data1 = data1.dropna(subset=["Variación Ranking Tamaño 2020 - 2021"])
data1["Variación Ranking Tamaño 2020 - 2021"] = data1["Variación Ranking Tamaño 2020 - 2021"].astype(int)

menor = data1.loc[data1['Variación Ranking Tamaño 2020 - 2021'] < 1].count()[0]
mayor = data1.loc[data1['Variación Ranking Tamaño 2020 - 2021'] >= 1].count()[0]
print(menor)
print(mayor)
```

```python
 
etiquetas = ['Empeorado', 'Mejorado']
valores = [menor, mayor]
plt.figure()
barras = plt.bar(etiquetas, valores)
plt.show()
```

## 2. Generar gráfico de las empresas que tienen panamá papers

```python
data = df.loc[df['No. de Integrantes y Empresas Offshore relacionadas con el Grupo Económico publicados en PANAMÁ PAPERS (***)'] > 0]

data = data.drop([303])

nombres=[]
for valor in data["\nNombre Grupo Económico"]:
  nombres.append(valor)

plt.bar(data['\nNombre Grupo Económico'], data['No. de Integrantes y Empresas Offshore relacionadas con el Grupo Económico publicados en PANAMÁ PAPERS (***)'], color='yellow')
plt.show
```

## 3. Generar un pie de las empresas que más aportan a la economía del país.

```python
data3 = pd.DataFrame(df, columns=['\nNombre Grupo Económico', 'Ranking Contribución I.R. 2020', 'Contribución   I.R. 2020'])

data3.head(10)
```

```python
import re

contribucion=data3.loc[data3["Ranking Contribución I.R. 2020"] <=10]
listValor=[]
for valor in contribucion["Contribución   I.R. 2020"]:
  listValor.append(float(valor))

ets=[]
for nombres in contribucion["\nNombre Grupo Económico"].values:
  ets.append(nombres.split("(")[0])
plt.pie(listValor, labels=ets, autopct='%.2f %%')
plt.show()
```

## 4. Obtener las empresas que mayor patrimonio tienen (10).

```python
df['Total Patrimonio (7) 2020']  = df['Total Patrimonio (7) 2020'].replace(0,np.nan)
patrimonio = df.sort_values(['Total Patrimonio (7) 2020'],ascending=False)
nuevospatrimonios=patrimonio['Total Patrimonio (7) 2020']
empresas=df['\nNombre Grupo Económico']

plt.bar(empresas[1:11], nuevospatrimonios[1:11], color='yellow')
plt.title("Mayor Patrimonio")
plt.show()
```
