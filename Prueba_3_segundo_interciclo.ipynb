# Planteamineto del Problema

En una biblioteca se necesita describir la cantidad de libros prestados a la semana. Segun los datos que tienen registrados la biblioteca tenemos que, en cada semana se prestan un minimo de 4 libros y un maximo de 8, pero en la mayoria de las semanas se prestan 6.

Para la simulacion de distribucion vamos a usar el valor mas recurrente sobre la prestacion de libros, que es 6, como valor de a.

# Simulacion

```python
#Librerias necesarias
import random
import math
import matplotlib.pyplot as plt
```

```python
#Generador de numeros aleatorios

def random_python(N=100):
  numeros = []
  random.seed(1032022)
  [numeros.append(random.random())for i in range(N)]
  return numeros

valoresAleatorios = []
valoresAleatorios = random_python()

print("Valores : ",  valoresAleatorios)
print(len(valoresAleatorios))
```

```python
# Aplicacion de la Funcion Triangular
a = 6.00
contador = 0
valor=0
valorPresLibro = []
semana= []
print("Num. de Semana  \t ri \t probabilidad de prestación")
for i in valoresAleatorios:
  valor = a*(1- math.sqrt(1 - i))
  valorPresLibro.append(valor)
  contador = contador + 1
  semana.append(contador)
  print(str(contador) + "\t" + str(i) + "\t"+ str(valor))


# Funcion de Densidad
valorDensidad = []
contador2 = 0
print("Num. de Semana  \t ri \t Valor de Densidad")
for i in valoresAleatorios:
  valorD = 2/a * (1 - i/a)
  #valorD = 2/a * (i - i**2/(2*a))
  valorDensidad.append(valorD)
  contador2 = contador2 + 1
  print(str(contador2) + "\t" + str(i) + "\t"+ str(valorD))

```

# Gráficas/Resultados





```python
x = semana
y = valorPresLibro

plt.figure()
#plt.figure(figsize=(6,2), dpi=300)
plt.title('Grafica de Probabilidad Acumulada')
plt.xlabel('Num. de Semana')
plt.ylabel('Prestación')
plt.plot(x,y, color='red')
plt.show()
```

```python
x = semana
y = valorDensidad

plt.figure()
#plt.figure(figsize=(6,2), dpi=300)
plt.title('Grafica de Densidad')
plt.xlabel('Num. de Semana')
plt.ylabel('Densidad')
plt.plot(x,y, color='red')
plt.show()
```
