**Ejercicio en clases: Calcular, graficar y obtener el valor de Chi Cuadrado para los siguientes generadores con las siguientes semillas:**

Cuadrados medios: Xo=74731897457, D=7

Congruencia lineal: a=74731897457, b=37747318974, Xo=7, M=19

Tauswoth q=7, r=3, l=5

Random de Python semilla (11052022)

```python
# Librerias necesarias

import matplotlib.pyplot as plt
import random
import pandas as pd
import numpy as np
```

```python
# Cuadrados medios

def cuadrados_medios():
    digitos = 7
    xn = 74731897457
    valoresRamdom = []

    for i in range(100):
        xnn = xn**2
        txnn = str(xnn).zfill(8)
        tam = len(txnn)
        ui = int(txnn[int(tam/2-digitos/2):int(tam/2+digitos/2)])
        rn = ui / (int('9'*digitos)+1)
        xn = ui
        valoresRamdom.append(rn)
    return valoresRamdom

#print("Valores Generados 1 : " , valoresRamdom)
```

```python
# Congruencia lineal

def congruencia_lineal():
    xo = 7
    a = 74731897457
    b = 37747318974
    m = 19
    iteraciones = 100
    valoresRamdom2 = []
    xn = xo
    
    for i in range(iteraciones):
        xnn = (a*xn+b) % m
        ui = xnn/m
        xn = xnn
        valoresRamdom2.append(ui)
        
    return valoresRamdom2

#print("Valores Generados 2 : ",  valoresRamdom2)
#print(len(valoresRamdom2))
```

```python
def random_python(N=100):
    numeros = []
    random.seed(11052022)
    [numeros.append(random.random())for i in range(N)]
    return numeros

#valoresRamdom3 = []
#valoresRamdom3 = random_python()

#print("Valores Generados 3 : ",  valoresRamdom3)
#print(len(valoresRamdom3))
```

```python
# Tauswoth

def tauswoth():
    
    valoresRamdom4 = []
    q=7
    h=3 #r
    l=5
    nBits=500 #700

    generador=[]
    conversion=[]
    conversion2=[]
    result=0

    vector_l=list(range(l))
    vector_l.reverse()

    for i in range(1,len(list(range(1,nBits+1)))+1):
        if i<=q:
            generador.append(1)
        else:
            aux1=i-(h+1)
            aux2=i-(q+1)
            generador.append((generador[aux1]+generador[aux2])%2)

    for i in range(l,len(generador)+1,l):
        fragmento=generador[i-l:i]
        for j in range(len(fragmento)):
            result+=fragmento[j]*2**vector_l[j]
        conversion.append(result)
        conversion2.append(result/(2**l))
        result=0

    df = pd.DataFrame(generador,index=list(range(1,nBits+1)))
    valoresRamdom4 = conversion2
    
    return valoresRamdom4
```

```python
# Chi-cuadrado

#Defino una funcion para realizar un conteo de los numeros que caen dentro de cada intervalo
def calcular_chi(numeros):
    tablas = {}
    for i in np.arange(0.1, 1.1, 0.1): # aqui separo en intervalos de 0.1
        for j in numeros:
            if j > i-0.1 and j <= i: # Compruebo si el valor esta dentro del intervalor
                tablas[round(i-0.1,1)] = tablas[round(i-0.1,1)]+1 if round(i-0.1,1) in tablas else 1
    return tablas

#defino una funcion para aplicar la formula y obtener la desviacion estandar de cada intervalo  
def sumatoria_chi(tabla, E):
    return sum([(valor-E)**2/E for valor in tabla.values()])
```

```python
# Graficas 

#Llamada a Cuadrados medios
cMedios = cuadrados_medios()
tabla1 = calcular_chi(cMedios)
suma1 = sumatoria_chi(tabla1, 10)

#Llamada a Congruencia lineal
cLineal = congruencia_lineal()
tabla2= calcular_chi(cLineal)
suma2 = sumatoria_chi(tabla2, 10)

#Llamada a Random Python
rPython = random_python()
tabla3= calcular_chi(rPython)
suma3 = sumatoria_chi(tabla3, 10)

#Llamada a Tauswoth
rtauswoth = tauswoth()
tabla4= calcular_chi(rtauswoth)
suma4 = sumatoria_chi(tabla4, 10)

fig, ax = plt.subplots(2, 2, sharey = False, sharex = False)
fig.set_size_inches(20,10)
ax[0,0].bar(range(len(tabla1)), list(tabla1.values()), tick_label=list(tabla1.keys()))
ax[0, 0].set_title("Cuadrados Medios")

ax[0,1].bar(range(len(tabla2)), list(tabla2.values()), tick_label=list(tabla2.keys()))
ax[0, 1].set_title("Congruencia lineal")

ax[1,0].bar(range(len(tabla3)), list(tabla3.values()), tick_label=list(tabla3.keys()))
ax[1, 0].set_title("Random Python")

ax[1,1].bar(range(len(tabla4)), list(tabla4.values()), tick_label=list(tabla4.keys()))
ax[1, 1].set_title("Tauswoth")

plt.show()

print("Cuadrados Medios")
print("Tabla de conteo")
print(tabla1)
print("Valor del Chi-cuadrado")
print(suma1)
print("")

print("Congruencia Lineal")
print("Tabla de conteo")
print(tabla2)
print("Valor del Chi-cuadrado")
print(suma2)
print("")

print("Random Python")
print("Tabla de conteo")
print(tabla3)
print("Valor del Chi-cuadrado")
print(suma3)
print("")

print("Tauswoth")
print("Tabla de conteo")
print(tabla4)
print("Valor del Chi-cuadrado")
print(suma4)
print("")

```
